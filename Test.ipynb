```python
Check the goodness of fit for the following data mapping number of hours a student studies v/s the expected number of hours at 5% LOS .
observed data in hours - 8 6 10 7 8 11 9
expected data in hours - 9 8 11 8 10 7 6 
```

```python
observed = [8, 6, 10, 7, 8, 11, 9]
expected = [9, 8, 11, 8, 10, 7, 6]
n=len(observed)
chi_square = 0

print("O  E   (O-E)^2/E")

for O, E in zip(observed, expected):
    value = ((O - E) ** 2) / E
    chi_square += value
    print(O, "\t", E, "\t", round(value, 3))

print("Chi-Square Value =", round(chi_square, 3))
df = n - 1
print("Degree of Freedom =", df)

critical_value = 12.592
print("Critical Value =", critical_value)

if chi_square < critical_value:
    print("Accept Null Hypothesis")
    print(" The data shows Goodness of Fit")
else:
    print("Reject Null Hypothesis")
    print(" The data does not show Goodness of Fit")
```

```python
What is the Probability of 2 heads in 6 coin tosses? fit the binomial distribution for the coin tosses and plot it.

    
```

```python
import math
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 6   
p = 0.5 
k = 2   

prob_2_heads = binom.pmf(k, n, p)
print(f"Probability of exactly 2 heads in 6 tosses: {prob_2_heads:.4f}")

x = range(n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution: 6 Coin Tosses (p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(x)
plt.show()
```
