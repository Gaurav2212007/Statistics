```python
#Q1.a

import math
from scipy import stats

n1, mean1, var1 = 13, 15, 87.5
n2, mean2, var2 = 11, 12, 39

sp = ((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2)
se = math.sqrt(sp * (1/n1 + 1/n2))

t_stat = (mean1 - mean2) / se
df = n1 + n2 - 2

p_value = stats.t.sf(abs(t_stat), df)*2

print("t-stat:", t_stat, "p-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis: Means differ significantly")
else:
    print("Fail to Reject Null Hypothesis: No significant difference")
```

```python
#Q2.
# ANOVA AND TUKEY HSD 

from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

group1 = [24.5,23.5,26.4,27.1,29.9]
group2 = [28.4,34.2,29.5,32.2,30.1]
group3 = [26.1,28.3,24.3,26.2,27.8]

f_stat,p = f_oneway(group1,group2,group3)

print("F-statistics =", f_stat)
print("p-value =",p)

values = group1 + group2 + group3
groups = ['g1']*5 + ['g2']*5 + ['g3']*5

result = pairwise_tukeyhsd(values,groups)

print(result)
```
